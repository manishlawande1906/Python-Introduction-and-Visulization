```python
from math import *

plt.figure(figsize = (16,9))             ## fig size of x & y

x = list(range(-10,11))

y = [i**2 for i in x]
plt.subplot(3,4,1)                        ##figure has 3 rows & 4 columns & this is 1st graph
plt.plot(x,y)
plt.title("Graph of x**2")

y = [i**3 for i in x]
plt.subplot(3,4,10)                       ##figure has 3 rows & 4 columns & this is 10th graph
plt.plot(x,y)
plt.title("Graph of x**3")

y = [sin(i) for i in x]
plt.subplot(3,4,8)
plt.plot(x,y)
plt.title("Graph of sin(i)")
```

```python
from math import *

plt.figure(figsize = (16,4))             ## fig size of x & y

plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.title('Boxplot of tips')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.title('Violin plot of tips')

plt.show()
```
